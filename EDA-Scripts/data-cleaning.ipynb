Things to remove: 0 BPM, 0 duration, duplicates

```python
import pandas as pd

df = pd.read_csv('../datasets/tracks.csv')
print(df.columns)
df.head()
```

```python
print(f"Rows with 0 BPM (Tempo): {(df['tempo'] == 0).sum()}")
print(f"Rows with 0 Time Signature: {(df['time_signature'] == 0).sum()}")
#track names shouldn't matter too much, b/c it's not rly a numeric feature
print(f"Missing track names: {df['name'].isna().sum()}")
print(f"Tracks shorter than 10 seconds: {(df['duration_ms'] == 0).sum()}")
print(f"Duplicate Name + Artist pairs: {df.duplicated(subset=['name', 'artists']).sum()}")
print(f"Rows with 0 Duration: {(df['duration_ms'] == 0).sum()}")
print(f"Rows with 0 Danceability: {(df['danceability'] == 0).sum()}")
print(f"Rows with 0 Energy: {(df['energy'] == 0).sum()}")
print(f"Rows with 0 Speechiness: {(df['speechiness'] == 0).sum()}")
print(f"Rows with 0 Acousticness: {(df['acousticness'] == 0).sum()}")
print(f"Rows with 0 Instrumentalness: {(df['instrumentalness'] == 0).sum()}")
print(f"Rows with 0 Liveness: {(df['liveness'] == 0).sum()}")
print(f"Rows with 0 Valence: {(df['valence'] == 0).sum()}")
```

Seems like crazy amount of dupe songs

```python
dupe_ct = df.groupby(['name', 'artists']).size().reset_index(name='count')
print(dupe_ct.sort_values('count',ascending=False).head(10))
```

```python
bieber_hold_on = df[(df['name'] == 'Hold On') & (df['artists'].str.contains('Justin Bieber'))]
print(f"Found {len(bieber_hold_on)} versions of 'Hold On' by Justin Bieber:")
pd.set_option('display.max_rows', None)  # Ensure all rows are printed
print(bieber_hold_on[['id', 'name', 'artists', 'popularity', 'release_date', 'duration_ms']])
```

```python
noInstruments = df[(df['instrumentalness']==0)]
print(noInstruments[['name', 'artists']])
```

```python
noDance = df[(df['danceability']==0)]
print(noDance[['name', 'artists']])
```

```python
df_cleaned = df[
    (df['tempo'] > 0) & 
    (df['time_signature'] > 0) & 
    (df['duration_ms'] >= 10000)
]
df_cleaned = df_cleaned.sort_values('popularity', ascending=False).drop_duplicates(subset=['name', 'artists'])
print(f"Original Row Count: {len(df)}")
print(f"Cleaned Row Count:  {len(df_cleaned)}")
print(f"Rows Removed:       {len(df) - len(df_cleaned)}")
df_cleaned.to_csv('../datasets/tracks_cleaned.csv', index=False)
print("\n file saved as 'tracks_cleaned.csv'")
```
