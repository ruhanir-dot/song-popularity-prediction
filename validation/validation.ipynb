```python
import joblib
import pandas as pd
from sklearn.metrics import classification_report

# OHE key and mode
new_ohe = pd.get_dummies(new_df, columns=['key', 'mode'])

# OHE explicit manually (to match training format)
new_ohe['explicit_0'] = (new_df['explicit'] == 0).astype(int)
new_ohe['explicit_1'] = (new_df['explicit'] == 1).astype(int)

# Load model and columns
model = joblib.load('best_rf_model.pkl')
model_columns = joblib.load('model_columns.pkl')

# Align to training columns exactly
new_X = new_ohe.reindex(columns=model_columns, fill_value=0)

# Verify shape looks right
print(f"new_X shape: {new_X.shape}")
print(f"Any nulls: {new_X.isnull().sum().sum()}")

# Bucket actual popularity for comparison
new_df['actual_bucket'] = pd.cut(
    new_df['popularity'],
    bins=[0, 25, 50, 100],
    labels=['Low', 'Medium', 'High']
)

# Predict
predictions = model.predict(new_X)
new_df['predicted_bucket'] = predictions

# Results
valid = new_df['actual_bucket'].notna()
print(classification_report(
    new_df.loc[valid, 'actual_bucket'],
    new_df.loc[valid, 'predicted_bucket']
))

# Spot check
print(new_df[['track_name', 'track_artist', 'popularity',
              'actual_bucket', 'predicted_bucket']].head(20))
```
